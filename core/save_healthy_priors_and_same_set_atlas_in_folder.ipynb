```python
from cProfile import label
from email.mime import image
import os 
import csv
import shutil
import sys
os.chdir('/home/malek/mock/autoDDPM')
dir_path = os.path.dirname(os.path.realpath(os.getcwd()))


# Get the parent directory
# Add the parent directory to the system path
sys.path.append(dir_path)
print(dir_path)

image_folder = f"/home/malek/mock/autoDDPM/data/ATLAS/splits_healthy_54_sample/Atlas_healthy.csv"

output_folder = f"/home/malek/mock/autoDDPM/data/ATLAS/splits_healthy_54_sample/images"
os.makedirs(output_folder, exist_ok=True)


# Open the CSV file
with open(image_folder, 'r') as file:
    # Create a CSV reader object
    csv_reader = csv.reader(file)
    
    # Iterate over each line in the CSV file
    for line in csv_reader:
        # Get the path to the image
        image_path = line[0]

        if image_path == 'filename':
            continue
        print('image path', image_path)
        
        # Extract the image name from the path
        image_name = os.path.basename(image_path)
        
        # Create the output path for the image
        output_path = os.path.join(output_folder, image_name)
        
        # Copy the image to the output folder
        shutil.copy(image_path, output_path)
```
