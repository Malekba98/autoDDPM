```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def print_predictions_train_set(index):
    # Load the image

    img = mpimg.imread(f'/home/malek/autoDDPM/nnunet_data/nnunet_results/Dataset500_ATLAS/nnUNetTrainer__nnUNetPlans__2d/crossval_results_folds_0_1_2_3_4/ATLAS_{index}.png')

    # Create a figure with two subplots
    fig, axs = plt.subplots(1, 3, figsize=(6, 3))

    # Display the first image
    axs[0].imshow(img)
    axs[0].set_title('predicted mask')
    axs[0].axis('off')

    # Load the second image
    img2 = mpimg.imread(f'/home/malek/autoDDPM/nnunet_data/nnunet_raw/Dataset500_ATLAS/labelsTr/ATLAS_{index}.png')

    # Display the second image
    axs[1].imshow(img2)
    axs[1].set_title('original mask')
    axs[1].axis('off')


    img3 = mpimg.imread(f'/home/malek/autoDDPM/nnunet_data/nnunet_raw/Dataset500_ATLAS/imagesTr/ATLAS_{index}_0000.png')
    #axs[2].imshow(img3,cmap='gray')
    #axs[2].set_title('Brain image')
    #axs[2].axis('off')


    # Adjust the spacing between subplots

    # Plot the brain image
    axs[2].imshow(img3, cmap='gray')
    axs[2].set_title('Brain image')
    axs[2].axis('off')

    # Plot the contours of the predicted mask
    axs[2].contour(img, levels=[0.5], colors='red', alpha=0.5)

    # Plot the contours of the original mask
    axs[2].contour(img2, levels=[0.5], colors='blue', alpha=0.5)


    # Adjust the spacing between subplots
    plt.tight_layout()

    # Show the figure
    plt.show()
```
