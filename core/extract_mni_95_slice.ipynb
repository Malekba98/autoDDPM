```python
import nibabel as nib
import numpy as np
import numpy as np
import matplotlib.pyplot as plt

# Specify the file path
file_path = '/home/malek/autoDDPM/data/MNI152_template/icbm_avg_152_t1_tal_nlin_symmetric_VI.nii'

# Load the NIfTI file
nifti_img = nib.load(file_path)

# Access the data and header information
data = nifti_img.get_fdata()
header = nifti_img.header

print('Data shape:', data.shape)

# Transpose the data
data = np.transpose(data, (1, 0, 2))

# Display the transposed data shape
print('Transposed data shape:', data.shape)

# Turn the data upside down
data = np.flipud(data)

# Display the transposed data shape
print('Upside down data shape:', data.shape)

# Do further processing with the NIfTI data
# ...
# Extract the 95th slice from the axial view
axial_slice = data[:, :, 94]

print('type of axial_slice:', type(axial_slice))
print('shape of axial_slice:', axial_slice.shape)

import matplotlib.pyplot as plt

# Save the axial_slice array as a .png file
plt.imsave('axial_slice.png', axial_slice, cmap='gray')



# Display the extracted slice
plt.imshow(axial_slice, cmap='gray')
plt.axis('off')

plt.show()
```
